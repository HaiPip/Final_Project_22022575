```python
import pandas as pd
```

```python
data_clean = pd.read_csv('Data/TamthanWibu.csv')
```

```python
# Xóa cột có tất cả giá trị là Null
data_clean = data_clean.dropna(axis=1, how='all')
```

```python
print (data_clean.columns)
```

```python
# lấy biểu cảm của từng bài đăng
def split_reactions(row):
    reaction_dict = eval(row['reactions'])
    return pd.Series(reaction_dict)
reactions = data_clean.apply(split_reactions, axis=1)
```

```python
reactions
```

```python
# Đổi giá trị null sang 0
reactions = reactions.fillna(0)
reactions = reactions.astype('int64')
```

```python
reactions
```

```python
data_clean
```

```python
# liên kết data
data_clean = pd.concat([data_clean, reactions], axis=1)
data_clean.head()
```

```python
# đổi kiểu dữ liệu fetched_time thành datetime64
data_clean['time'] = pd.to_datetime(data_clean['time'])
data_clean['time']
```

```python
# lấy thời gian đăng của mỗi bài
data_clean['hour'] = data_clean['time'].dt.hour
data_clean['day'] = data_clean['time'].dt.day
data_clean['month'] = data_clean['time'].dt.month
data_clean['year'] = data_clean['time'].dt.year
data_clean[['hour', 'day', 'month', 'year']]
```

```python
data_clean.to_csv('Data/data_clean.csv', index=False)
```
